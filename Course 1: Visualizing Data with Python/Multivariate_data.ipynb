Since the waist measurement column contains 'NaN' values, using np.median we lead to an error.

Hence, the alternate options are:

- Using the median method from pd library, which excludes the 'NaN' values
- Using .nanmedian method from np numpy, which also excludes the 'NaN' values

```python
# waist measurements larger than the median

waist_median = pd.Series.median(df['BMXWAIST'])

waist_median
```

```python
condition1 = df["BMXWAIST"] > waist_median

condition1
```

```python
df[condition1].head()
```

Using the condition of waist measurements being greater than waist median, we obtain a data frame containing rows which fulfill the condition.

```python
condition2 = df['BMXLEG'] < 32

df[condition1 & condition2].head()

# notice we do not use 'and' keyword
# instead we use '&' symbol 
```

# Multivariate Distributions in Python
---

Sometimes, we can get a lot of information about how two variables (or more) relate if we plot them together. 

```python
import matplotlib.pyplot as plt
import seaborn as sns

r = 0.6

mean = [15,5]
cov = [[1,r],[r,1]]
x, y = np.random.multivariate_normal(mean,cov,400).T
```

```python
plt.figure(figsize = (10,5))

plt.subplot(1,2,1)
plt.hist(x=x,bins=15)
plt.title("X")

plt.subplot(1,2,2)
plt.hist(x=y,bins=15)
plt.title("Y")
```

```python
plt.figure(figsize=(10,10))
plt.subplot(2,2,2)
plt.scatter(x=x,y=y)
plt.title("Joint Distribution of X and Y")

# Plot the Marginal Distribution of X
plt.subplot(2,2,4)
plt.hist(x=x,bins=15)
plt.title("Marginal Distribution of X")

# Plot the Marginal Distribution of Y
plt.subplot(2,2,1)
plt.hist(x=x,bins=15,orientation='horizontal')
plt.title("Marginal Distribution of Y")
```

The basic idea that revolves around Unit Testing is running tests on a small copy of the dataset to check the procedures and functions work correctly.

On a large dataset, it might not be possible to check the answer. 

Hence, testing the code on a smaller dataset, where you can add test values is an useful method to check whether you get the answer that you expect.
